# Document clusters & gold-label comparison (all models)

Runs over **all** run folders under `data_out/data_out`. For each model: (1) % per cluster, (2) % per gold label, (3) bar charts. No embeddings.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

DATA_ROOT = Path("../data_out/custom_data")
RUN_INDEX = 0

def find_run_folders(root: Path):
    return sorted(set(p.parent for p in Path(root).rglob("document_assignments.csv")))

run_folders = find_run_folders(DATA_ROOT)
print(f"Found {len(run_folders)} models under {DATA_ROOT}")
if not run_folders:
    raise SystemExit("No run folders found. Run an experiment first.")
```

```python
# For each model: Block 1 (cluster %), Block 2 (gold %), Block 3 (graph)
for RUN_FOLDER in run_folders:
    df = pd.read_csv(RUN_FOLDER / "document_assignments.csv")
    df_run = df[df["run"] == RUN_INDEX].reset_index(drop=True)
    n_docs = len(df_run)
    name = RUN_FOLDER.name

    # Block 1: Percentage per cluster (assigned topic)
    assigned = df_run["llm_assigned_topic"].value_counts().sort_index()
    print(f"\n{'='*60}\n{name} | {n_docs} docs\n{'='*60}")
    print("Percentage of docs per cluster (assigned topic):\n")
    for topic, count in assigned.items():
        pct = 100 * count / n_docs
        print(f"  {topic[:55]}{'...' if len(str(topic)) > 55 else ''}: {pct:.1f}% ({count})")

    # Block 2: Percentage per gold label
    gold = df_run["original_category"].value_counts().sort_index()
    print("\nPercentage of docs per gold label:\n")
    for label, count in gold.items():
        pct = 100 * count / n_docs
        print(f"  {label}: {pct:.1f}% ({count})")

    # Block 3: Graph — bar charts for cluster % and gold label %
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    pct_assigned = (assigned / n_docs * 100).sort_values(ascending=True)
    axes[0].barh(range(len(pct_assigned)), pct_assigned.values)
    axes[0].set_yticks(range(len(pct_assigned)))
    axes[0].set_yticklabels([str(t)[:40] + ("." if len(str(t)) > 40 else "") for t in pct_assigned.index], fontsize=7)
    axes[0].set_xlabel("% of documents")
    axes[0].set_title("Assigned topic (cluster)")
    pct_gold = (gold / n_docs * 100).sort_values(ascending=True)
    axes[1].barh(range(len(pct_gold)), pct_gold.values)
    axes[1].set_yticks(range(len(pct_gold)))
    axes[1].set_yticklabels(pct_gold.index, fontsize=8)
    axes[1].set_xlabel("% of documents")
    axes[1].set_title("Gold label")
    fig.suptitle(name, y=1.02)
    plt.tight_layout()
    plt.show()
```
