## Coherence scores: mean and standard deviation
Set the path to your coherence CSV below, then run all cells.

```python
import pandas as pd

# Set path to your coherence_scores_*.csv file
COHERENCE_CSV_PATH = "data_out/GenAIMethodOneShot_400_NYT_gpt-5.2_equal/coherence_scores.csv"
```

```python
df = pd.read_csv(COHERENCE_CSV_PATH)
stats = df.groupby("metric_name")["score"].agg(["mean", "std"]).reset_index()

metrics_order = ["V_measure", "completeness", "homogeneity"]
parts = []
for m in metrics_order:
    row = stats[stats["metric_name"] == m].iloc[0]
    mean_val = row["mean"]
    sd_val = row["std"] if pd.notna(row["std"]) else 0.0
    parts.append(f"{mean_val:.2f} ({sd_val:.3f})")
print(" ".join(parts))
```
